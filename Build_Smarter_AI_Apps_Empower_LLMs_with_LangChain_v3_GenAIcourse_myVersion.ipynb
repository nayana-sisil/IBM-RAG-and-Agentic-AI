# **loading the model**

```python
import os
from langchain.chat_models import init_chat_model


model = init_chat_model(
    "deepseek-ai/DeepSeek-R1-Distill-Qwen-1.5B",
    model_provider="huggingface",
    temperature=0.7,
    max_tokens=1024,
)
```

```python
response = model.invoke("Why do parrots have colorful feathers?")
print(response)
```

```python
response = model.invoke("what is  4  plus 4")
print(response)
```

# **Chat message**

```python
from langchain.messages import HumanMessage, AIMessage, SystemMessage

conversation = [
    SystemMessage("You are a helpful assistant that tell me about today job market honestly"),
    HumanMessage("tell me about software job market"),
    AIMessage("it's abousulte mess"),
    HumanMessage("what about data science  job market?")
]

response = model.invoke(conversation)
print(response)
```

```python
print(response.content)
```

# **Prompt templates**

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate.from_template(
    "Tell me about the {field} job market in 2026."
)

prompt = template.invoke({"field": "data science"})
print(prompt)
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template(
    "Tell me about the {field} job market."
)

chain = prompt | model

response = chain.invoke({"field": "software"})
print(response.content)
```

```python
import requests
from langchain_community.document_loaders import PyPDFLoader

url = "https://arxiv.org/pdf/2005.11401.pdf"

response = requests.get(url)
with open("paper.pdf", "wb") as f:
    f.write(response.content)

loader = PyPDFLoader("paper.pdf")
documents = loader.load()

print(len(documents))
print(documents[0].page_content[:500])
```

# **Text splitters**

RecursiveCharacterTextSplitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    add_start_index=True,
)

all_splits = text_splitter.split_documents(documents)

print(f"Split blog post into {len(all_splits)} sub-documents.")
```

# **Embedding models**

```python
!pip install langchain langchain-community sentence-transformers
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

text = "Hello world"
vector = embeddings.embed_query(text)

print(len(vector))
print(vector[:10])
```

```python
texts = [text.page_content for text in all_splits]

embedding_result = embeddings.embed_documents(texts)
embedding_result[0][:5]
```

# **Vector stores**

```python
from langchain_community.vectorstores import Chroma
```

```python
!pip install chromadb
```

```python
docsearch = Chroma.from_documents(all_splits, embeddings)
```

```python
query = "Large langagee models"
docs = docsearch.similarity_search(query)
print(docs[0].page_content)
```

# Vector store-backed retrievers

```python
retriever = docsearch.as_retriever()


docs = retriever.invoke("Langchain")


docs[0]
```
